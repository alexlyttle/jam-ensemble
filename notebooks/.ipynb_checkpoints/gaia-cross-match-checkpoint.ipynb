```python
from astropy.table import Table
from astropy.io import fits
import pandas as pd
import numpy as np
import os
```

```python
PATH_TO_SEIS = '/Users/ajl573/Repos/tkallinger/KeplerRGpeakbagging/summary.dat'  # TXT format
PATH_TO_KIC_GAIA = '/Users/ajl573/Data/kepler_dr2_4arcsec.fits'  # FITS format

SEIS_ID_COL = 'ID'
KIC_GAIA_ID_COL = 'kepid'

OUTPUT_DIR = '/Users/ajl573/Data'
OUTPUT_PREFIX = 'tkallinger_xmatch_'
```

```python
df_seis = pd.read_csv(PATH_TO_SEIS, delimiter=';')
print(len(df_seis))
df_seis.head()
```

```python
kic_gaia_fits = fits.open(PATH_TO_KIC_GAIA)
```

```python
kic_gaia_table = Table(kic_gaia_fits[1].data)
```

```python
df_kic_gaia = kic_gaia_table.to_pandas()
print(len(df_kic_gaia))
df_kic_gaia.head()
```

```python
df_cross = df_seis.merge(df_kic_gaia, left_on=SEIS_ID_COL, right_on=KIC_GAIA_ID_COL)
print(len(df_cross))
df_cross.head()
```

```python
df_dups = df_cross[df_cross.duplicated('ID', keep=False)]
print(len(df_dups))
df_dups.head()
```

```python
duplicated_ids = list(set(df_dups['ID']))
```

```python
df_cross['duplicated'] = np.zeros(len(df_cross)).astype(int)
for dup_id in duplicated_ids:
    df_cross.loc[df_cross['ID'] == dup_id, 'duplicated'] = 1
df_cross.head()
```

```python
(df_cross['phot_g_mean_mag'] - df_cross['kmag']).plot(kind='hist')
```

```python
df_cross['nearest'] = np.zeros(len(df_cross)).astype(int)
for dup_id in duplicated_ids:
    nearest = df_cross.loc[df_cross['ID'] == dup_id, 'kepler_gaia_ang_dist'].idxmin()
    df_cross.loc[nearest, 'nearest'] = 1
```

```python
# Non-duplicated targets are also flagged as 'nearest'
df_cross.loc[~df_cross.duplicated('ID', keep=False), 'nearest'] = 1
```

```python
df_cross['gmag_kmag_gtr_5'] = np.zeros(len(df_cross)).astype(int)
condition = (df_cross['nearest'] == 1) & (df_cross['phot_g_mean_mag'] - df_cross['kmag'] > 5.0)
df_cross.loc[condition, 'gmag_kmag_gtr_5'] = 1
```

```python
df_cross.head()
```

```python
df_nearest = df_cross[df_cross['nearest'] == 1]
print(len(df_nearest))
df_nearest.head()
```

```python
if len(df_nearest[df_nearest['gmag_kmag_gtr_5'] == 1]) == 0:
    print('Gaia G-band and Kepler K-band magnitudes differ by no more than 5.0 dex in nearest neighbours.')
```

```python
df_nearest.to_csv(os.path.join(OUTPUT_DIR, OUTPUT_PREFIX+'nearest.csv'), index=False)
```

```python
df_cross.to_csv(os.path.join(OUTPUT_DIR, OUTPUT_PREFIX+'full.csv'), index=False)
```
